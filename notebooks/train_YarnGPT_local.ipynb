```python
from torch.utils.data import DataLoader, Dataset
class YarnDataset(Dataset):
  def __init__(self,dataset):
    self.ds = dataset
    super().__init__()

  def __len__(self):
    return len(self.ds)


  def __getitem__(self, idx):
    prompt=self.ds[idx]["tts"]
    #print(prompt)
    return tokenizer(prompt,)
```

```python
yarn_dataset = YarnDataset(train_dataset)
```

```python
batch_size=4
learning_rate=1e-3
```

```python
# Initialize data collator
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# Create DataLoader with collate_fn using data collator
dataloader = DataLoader(
    yarn_dataset,
    batch_size=batch_size,
    collate_fn=data_collator,shuffle=True  # Automatically handles padding
)
```

```python
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from torch.optim.lr_scheduler import LambdaLR
from transformers import get_linear_schedule_with_warmup,get_cosine_schedule_with_warmup,get_constant_schedule_with_warmup
```

```python
def get_lr_lambda(step):
    if step < lr_warmup_steps:
        # Linear warmup
        return step / lr_warmup_steps
    elif step >=(num_decay_start):
        return 1-(step-num_decay_start)/(num_training_steps-num_decay_start)
    else:
        # Constant learning rate
        return 1
```

```python
num_epochs=2
optimizer = AdamW(model.parameters(), lr=learning_rate, betas=(0.9, 0.95),weight_decay=0.01)
lr_warmup_steps=200

num_training_steps=1255*num_epochs
num_decay_start=50#num_training_steps#-20
#scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=T_0, T_mult=T_mult, eta_min=eta_min)
#scheduler = # Create LambdaLR scheduler
scheduler = get_constant_schedule_with_warmup(optimizer,num_warmup_steps=lr_warmup_steps)#LambdaLR(optimizer, lr_lambda=get_lr_lambda)    #get_constant_schedule_with_warmup(optimizer,num_warmup_steps=10)#
global_step = 0
accumulation_steps = int(512/batch_size)#32
model.train()
```

```python
for epoch in range(num_epochs):
    t0=time.time()
    loss_accum=0
    #batch_iterator = tqdm(data, desc=f"Processing Epoch {epoch:02d}")
    for i,batch in enumerate(dataloader):
        input_ids = batch['input_ids'].to(device).long()
        attention_mask = batch['attention_mask'].to(device).long()
        with torch.autocast(device_type="cuda", dtype=torch.bfloat16):
          outputs = model(input_ids=input_ids,
                      attention_mask=attention_mask,
                      labels=input_ids)
          loss=outputs.loss
          loss = loss/ accumulation_steps
          loss_accum+=loss.detach()
        loss.backward()


        if (i + 1) % accumulation_steps == 0:
          #print(i)
          norm=torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
          optimizer.step()
          optimizer.zero_grad(set_to_none=True)
          scheduler.step()
          global_step += 1


          # Get and format the learning rate
          #lr_rate = scheduler.get_last_lr()[0]
          t1=time.time()
          dt=t1-t0
          logs={"loss": f"{loss_accum}", "step": global_step,"num_iter":batch_size*(i+1),"lr":scheduler.get_last_lr()[0],"time":f"{dt} Seconds","norm":norm.item()}
          print(logs)
          with open("/content/drive/MyDrive/YarnGPT_naij/logs.json", "a") as file:
              json.dump(logs, file)
          t0=time.time()
          loss_accum=0
        if (i>0) and (i + 1) % (2*8192)== 0:
          torch.save({
                       'epoch': epoch,
                        'model_state_dict': model.state_dict(),
                        'optimizer_state_dict': optimizer.state_dict(),
                        'scheduler_state_dict':scheduler.state_dict(),
                        'loss': loss,
                        'global_step':global_step
                  },f'/content/drive/MyDrive/YarnGPT_naij/{i*batch_size}_{epoch}xtraepoch.pt')


          #model.push_to_hub(new_checkpoint,private=False,commit_message=f"model {epoch} {(i+1)*batch_size}")
          model.train()
    optimizer.step()
    torch.save({
                       'epoch': epoch,
                        'model_state_dict': model.state_dict(),
                        'optimizer_state_dict': optimizer.state_dict(),
                        'scheduler_state_dict':scheduler.state_dict(),
                        'loss': loss,
                        'global_step':global_step
                  },f'/content/drive/MyDrive/YarnGPT_naij/final_{epoch}xtraepoch.pt')
model.push_to_hub(new_checkpoint,private=False,commit_message=f"final")
```
